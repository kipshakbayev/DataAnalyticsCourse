```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
#1

x = np.linspace(0,10)
y = np.exp(x)
plt.plot(x,y)
```

```python
#2

a = np.linspace(1,10)
b = np.log(a)
b1 = np.log2(a)
b2 = np.log10(a)

plt.plot(a,b)
plt.plot(a,b1)
plt.plot(a,b2)
```

```python
#3

fig = plt.figure()

c = np.linspace(0,1)
d = np.arcsin(c)
d1 = np.arccos(c)
d2 = np.arctan(c)


plt.plot(c,d,'r')
plt.plot(c,d1,'--g')
plt.plot(c,d2,'-.y')


```

```python
fig.savefig('arcsincostan.png')
```

```python
#4

ax = plt.axes() 

x = np.linspace(0,10)
y = np.sinh(x)

plt.plot(x,y)
plt.axis([-2,12,-1000,12000])
ax.set(xlabel=' ', ylabel='sinh(x)', title='  ')
```

```python
#5

fig, ax = plt.subplots(2,3)

ax[0][0].plot(a,b)
ax[0][1].plot(a,b1)
ax[0][2].plot(a,b2)

ax[1][0].plot(c,d)
ax[1][1].plot(c,d1)
ax[1][2].plot(c,d2)
```
