```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#1

sns.set_style('whitegrid', {'grid.color':'yellow'})

x = np.linspace(1,10)
y = np.log(x)
y1 = np.log2(x)
y2 = np.log10(x)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
#4-1

df3 = sns.load_dataset('diamonds')
df3
```

```python
#4-2

sns.scatterplot(x='carat', y='price', data=df3)
plt.axis('tight')
```

```python
#4-3

grid = sns.FacetGrid(df3, col='cut', hue='cut')
grid.map(sns.scatterplot, 'price', 'carat')
plt.axis('tight')
```

```python
#5-1

df4 = sns.load_dataset('car_crashes')
df4
```
